## 1. Classes and Objects - The Foundation of OOP

A **class** is like a blueprint for creating objects. An **object** is an instance of a class.

Think of a class as a cookie cutter (the template) and objects as the cookies (the actual instances).

```python
# Creating your first class
class Dog:
    # Class attribute (shared by all instances)
    species = "Canis familiaris"
    
    # Initializer method (constructor)
    def __init__(self, name, age):
        self.name = name  # Instance attribute
        self.age = age    # Instance attribute
    
    # Method (function inside a class)
    def bark(self):
        return f"{self.name} says Woof!"
    
    # Another method
    def get_human_years(self):
        return self.age * 7

# Creating objects (instances)
dog1 = Dog("Buddy", 3)
dog2 = Dog("Max", 5)

print(f"Dog 1: {dog1.name}, Age: {dog1.age}")
print(f"Dog 2: {dog2.name}, Age: {dog2.age}")
print(f"Species: {dog1.species}")
print(dog1.bark())
print(f"{dog2.name} is {dog2.get_human_years()} in human years")
```

### Understanding self

`self` refers to the current instance of the class. It's automatically passed when you call a method on an object.

```python
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year
        self.is_running = False
    
    def start_engine(self):
        self.is_running = True
        return f"{self.brand} {self.model} engine started!"
    
    def stop_engine(self):
        self.is_running = False
        return f"{self.brand} {self.model} engine stopped!"
    
    def get_info(self):
        status = "running" if self.is_running else "stopped"
        return f"{self.year} {self.brand} {self.model} - Engine: {status}"

# Create car objects
my_car = Car("Toyota", "Camry", 2022)
friend_car = Car("Honda", "Civic", 2021)

print(my_car.get_info())
print(my_car.start_engine())
print(my_car.get_info())
print(friend_car.get_info())
```

## 2. Object-Oriented Programming Concepts

### Encapsulation
Bundling data and methods that work on that data within one unit (class).

```python
class BankAccount:
    def __init__(self, account_number, owner, balance=0):
        self.account_number = account_number
        self.owner = owner
        self.__balance = balance  # Private attribute (note the double underscore)
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return f"Deposited ${amount}. New balance: ${self.__balance}"
        return "Invalid deposit amount"
    
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            return f"Withdrew ${amount}. New balance: ${self.__balance}"
        return "Invalid withdrawal amount"
    
    def get_balance(self):
        return f"Current balance: ${self.__balance}"

# Create a bank account
account = BankAccount("123456", "John Doe", 1000)
print(account.deposit(500))
print(account.withdraw(200))
print(account.get_balance())

# Trying to access private attribute directly (this won't work as expected)
# print(account.__balance)  # This would raise an error
```

### Inheritance
Creating new classes based on existing classes. The new class inherits attributes and methods from the parent class.

```python
# Parent class
class Animal:
    def __init__(self, name):
        self.name = name
        self.is_alive = True
    
    def eat(self):
        return f"{self.name} is eating"
    
    def sleep(self):
        return f"{self.name} is sleeping"

# Child class (inherits from Animal)
class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)  # Call parent constructor
        self.breed = breed
    
    def bark(self):
        return f"{self.name} barks: Woof!"
    
    # Override parent method
    def eat(self):
        return f"{self.name} the {self.breed} is eating dog food"

# Another child class
class Cat(Animal):
    def __init__(self, name, color):
        super().__init__(name)
        self.color = color
    
    def meow(self):
        return f"{self.name} meows: Meow!"

# Create instances
dog = Dog("Buddy", "Golden Retriever")
cat = Cat("Whiskers", "Orange")

print(dog.eat())  # Overridden method
print(dog.sleep())  # Inherited method
print(dog.bark())  # Dog-specific method
print()
print(cat.eat())  # Inherited method
print(cat.sleep())  # Inherited method
print(cat.meow())  # Cat-specific method
```

### Polymorphism
The ability to use the same interface for different data types or objects.

```python
class Bird:
    def __init__(self, name):
        self.name = name
    
    def make_sound(self):
        return f"{self.name} makes a sound"

class Sparrow(Bird):
    def make_sound(self):
        return f"{self.name} chirps: Tweet tweet!"

class Eagle(Bird):
    def make_sound(self):
        return f"{self.name} screeches: Screeee!"

class Penguin(Bird):
    def make_sound(self):
        return f"{self.name} honks: Honk honk!"

# Function that works with any bird (polymorphism)
def bird_sound(bird):
    return bird.make_sound()

# Create different birds
sparrow = Sparrow("Sparry")
eagle = Eagle("Eggy")
penguin = Penguin("Penny")

# Same function works with different types
print(bird_sound(sparrow))
print(bird_sound(eagle))
print(bird_sound(penguin))

# Using a list (polymorphism in action)
birds = [sparrow, eagle, penguin]
for bird in birds:
    print(bird_sound(bird))
```

## 3. Methods vs Functions

### Functions
Functions are independent blocks of code that perform a specific task.

```python
# Functions are defined outside of classes
def calculate_area(radius):
    """Function to calculate area of a circle"""
    return 3.14159 * radius ** 2

def greet(name):
    """Function to greet someone"""
    return f"Hello, {name}!"

# Using functions
area = calculate_area(5)
greeting = greet("Alice")

print(f"Area: {area}")
print(greeting)
```

### Methods
Methods are functions that are defined inside a class and operate on instances of that class.

```python
class Calculator:
    def __init__(self, value=0):
        self.value = value
    
    # This is a method - it operates on self
    def add(self, number):
        self.value += number
        return self.value
    
    def multiply(self, number):
        self.value *= number
        return self.value
    
    def reset(self):
        self.value = 0
        return self.value

# Create calculator instance
calc = Calculator(10)
print(f"Initial value: {calc.value}")

# Call methods
print(f"After adding 5: {calc.add(5)}")
print(f"After multiplying by 2: {calc.multiply(2)}")
print(f"After reset: {calc.reset()}")
```

### Key Differences

| Aspect | Functions | Methods |
|--------|-----------|---------|
| **Definition** | Independent of classes | Defined inside classes |
| **First Parameter** | No special first parameter | Always have `self` as first parameter |
| **Access to Object State** | Cannot access object attributes | Can access and modify object attributes |
| **Call Syntax** | `function_name(args)` | `object.method_name(args)` |
| **Purpose** | General purpose operations | Operations specific to the object |

```python
# Comparison example

# Function - independent
def get_circle_area(radius):
    return 3.14159 * radius ** 2

# Class with method
class Circle:
    def __init__(self, radius):
        self.radius = radius
    
    # Method - operates on instance
    def get_area(self):
        return 3.14159 * self.radius ** 2
    
    # Method can modify instance
    def set_radius(self, new_radius):
        self.radius = new_radius

# Using function
area1 = get_circle_area(5)
print(f"Area using function: {area1}")

# Using method
circle = Circle(5)
area2 = circle.get_area()
print(f"Area using method: {area2}")

# Method can change object state
circle.set_radius(10)
print(f"Area after changing radius: {circle.get_area()}")
```

## 4. Error Handling with try/except

Error handling allows your program to gracefully handle unexpected situations.

### Basic try/except
The basic structure for handling errors.

```python
# Example 1: Handling division by zero
try:
    numerator = 10
    denominator = 0
    result = numerator / denominator
    print(f"Result: {result}")
except ZeroDivisionError:
    print("Error: Cannot divide by zero!")
    print("Please provide a non-zero denominator.")

print("Program continues after handling the error...")
```

```python
# Example 2: Handling different types of errors
def safe_divide(a, b):
    try:
        result = a / b
        return f"Result: {result}"
    except ZeroDivisionError:
        return "Error: Cannot divide by zero!"
    except TypeError:
        return "Error: Both inputs must be numbers!"
    except Exception as e:
        return f"Unexpected error: {e}"

# Test the function
print(safe_divide(10, 2))      # Valid division
print(safe_divide(10, 0))      # Division by zero
print(safe_divide("10", 2))  # Type error
print(safe_divide(10, "2"))  # Type error
```

### try/except/else/finally
More comprehensive error handling structure.

```python
# Example with else and finally
def process_file(filename):
    try:
        file = open(filename, 'r')
        content = file.read()
        print(f"File '{filename}' opened successfully")
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found!")
        return None
    except Exception as e:
        print(f"Error reading file: {e}")
        return None
    else:
        # This runs only if no exception occurred
        print(f"File contains {len(content)} characters")
        return content
    finally:
        # This always runs, regardless of exceptions
        try:
            file.close()
            print("File closed")
        except:
            print("No file to close")

# Test the function
result = process_file("test.txt")
print()
result = process_file("nonexistent.txt")
```

### Exercise 1: Create a Student Class
Create a Student class with the following requirements:
- Attributes: name, student_id, grades (list)
- Methods: add_grade(), get_average(), get_status()
- get_status() should return "Pass" if average >= 60, otherwise "Fail"

```python
class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.grades = []

    def add_grade(self, grade):
        self.grades.append(grade)

    def get_average(self):
        return sum(self.grades) / len(self.grades)

    def get_status(self):
        if self.get_average() >= 60:
            return "Pass"
        else:
            return "Fail"


# Create a student
student1 = Student("John", "S001")

# Add grades
student1.add_grade(70)
student1.add_grade(80)
student1.add_grade(50)

# Display results
print("Name:", student1.name)
print("Student ID:", student1.student_id)
print("Grades:", student1.grades)
print("Average:", student1.get_average())
print("Status:", student1.get_status())
```

### Exercise 2: Error Handling for User Input
Create a function that safely gets user input and converts it to an integer:
- Keep asking until valid input is provided
- Handle ValueError for non-numeric input
- Handle KeyboardInterrupt when user presses Ctrl+C
- Return the valid integer

```python
def get_integer():
    while True:
        try:
            number = int(input("Enter an integer: "))
            return number

        except ValueError:
            print("Invalid input. Please enter a number.")

        except KeyboardInterrupt:
            print("\nProgram stopped by user.")
            return None


# Calling the function
result = get_integer()

if result is not None:
    print("You entered:", result)
```

### Exercise 3: Library Management System
Create a simple library system with error handling:
- Book class with title, author, is_available attributes
- Library class that manages multiple books
- Methods: add_book(), borrow_book(), return_book()
- Include appropriate error handling for:
  - Trying to borrow a book that doesn't exist
  - Trying to borrow a book that's not available
  - Trying to return a book that wasn't borrowed

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_available = True


class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)
        print(f'"{book.title}" added to the library.')

    def borrow_book(self, title):
        for book in self.books:
            if book.title == title:
                if book.is_available:
                    book.is_available = False
                    print(f'You borrowed "{book.title}".')
                    return
                else:
                    print("Error: Book is not available.")
                    return
        
        print("Error: Book does not exist.")

    def return_book(self, title):
        for book in self.books:
            if book.title == title:
                if not book.is_available:
                    book.is_available = True
                    print(f'You returned "{book.title}".')
                    return
                else:
                    print("Error: This book was not borrowed.")
                    return
        
        print("Error: Book does not exist.")


# Create books
book1 = Book("Things Fall Apart", "Chinua Achebe")
book2 = Book("Python Basics", "John Smith")

# Create library
library = Library()

# Add books
library.add_book(book1)
library.add_book(book2)

# Borrow books
library.borrow_book("Things Fall Apart")

# Try borrowing the same book again
library.borrow_book("Things Fall Apart")

# Try borrowing a book that does not exist
library.borrow_book("Data Science")

# Return a book
library.return_book("Things Fall Apart")

# Try returning a book that was not borrowed
library.return_book("Python Basics")
```
